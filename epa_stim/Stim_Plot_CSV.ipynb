```python
ind = np.arange(N)  # the x locations for the groups
width = 0.25  

ig, ax = plt.subplots()
plot1 = ax.bar(ind, rtpj_means, width, color='g', yerr=rtpj_sem)
plot2 = ax.bar(ind + width, rdlpfc_means, width, color='b', yerr=rdlpfc_sem)
plot3 = ax.bar(ind + width + width, sync_means, width, color='r', yerr=sync_sem)

ax.set_ylabel('% Rejected Difference')
ax.set_title('Rejections by Fairness Rating and Stimulation Type')
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('No Fairness Rating', 'Low Fairness Rating', 'High Fairness Rating'))
ax.legend((plot1[0], plot2[0], plot3[0]),('rTPJ', 'rDLPFC', 'Sync'), prop={'size': 6})
plt.locator_params(axis='y', nbins=5)
plt.ylim([0, 100])
def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d' % int(height),
                ha='center', va='bottom')
plt.show()
plt.savefig('tACS_plot.png')
```
